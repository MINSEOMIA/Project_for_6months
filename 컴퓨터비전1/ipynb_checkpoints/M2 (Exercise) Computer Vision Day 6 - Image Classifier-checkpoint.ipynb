## 전처리(Preprocessing)

### 데이터 적재

```python
import os
```

```python
def load_dataset(image_dir):
    image_list = []
    image_types = ["day", "night"]
    
    for image_type in image_types:
        file_paths = glob.glob(os.path.join(image_dir, image_type, "*.jpg"))
        for file_path in file_paths:
            image = mpimg.imread(file_path)
            image_list.append((image, image_type))    
    return image_list
```

```python
train_image_list = load_dataset(image_dir_training) # train_image_list에는 240개의 image/label list
test_image_list = load_dataset(image_dir_test) # test_image_list에는 160개의 image/label list
```

```python
print("train len:", len(train_image_list))
print("0번째 train image shape:", train_image_list[0][0].shape)
print("0번째 train image label:", train_image_list[0][1])
```

### 표준화 (Resize, Label Encoding)

```python
# 입력 이미지를 resize
def resize_input(image):
    resized_image = cv2.resize(image, (1200, 600)) # width:1200, height:600
    return resized_image

# label encoding
def labed_encode(label): # label <- day or night
    val = 0
    if label == "day":
        val = 1
    else:
        val = 0
    return val    
```

```python
def preprocess(image_list):
    preprocessed_list = []
    
    for item in image_list:
        image = item[0]
        label = item[1]
        
        processed_image = resize_input(image)
        processed_label = labed_encode(label)
        
        preprocessed_list.append([processed_image, processed_label])
        
    return preprocessed_list    
```

```python
preprocessed_train_list = preprocess(train_image_list)
preprocessed_test_list = preprocess(test_image_list)
```
